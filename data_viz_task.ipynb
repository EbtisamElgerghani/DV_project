```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('cars93.csv')
df
```

### 1. Boxplot

```python
# Boxplot for revs per mile
selected_manufacturers = ['Audi','Hyundai','Suzuki','Toyota']
filtered_df = df[df['Manufacturer'].isin(selected_manufacturers)]

plt.figure(figsize=(10, 6))
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=filtered_df)
plt.title('Revs per Mile by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Rev per Mile')
#plt.legend()
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Suzuki

### 2. Histogram

```python
# Histogram of MPG
plt.figure(figsize=(10, 6))
plt.hist(df['MPG.city'], alpha=0.5, label='City MPG', bins=15)
plt.hist(df['MPG.highway'], alpha=0.5, label='Hihgway MPG', bins=15)
plt.title('MPG Distribution')
plt.xlabel('MPG')
plt.ylabel('Frequancy')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
On the highway


### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure(figsize=(10, 6))
sns.lineplot(x='Wheelbase', y='Turn.circle', data=df)
plt.title('Wheelbase vs. Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
By looking to the Lineplot the Turning circle gets larger most of the times the wheelbase gets larger.

### 4. Barplot

```python
# Barplot for mean horsepower
mean_hoursepower_type = df.groupby('Type')['Horsepower'].mean().sort_values()
plt.figure(figsize=(10, 6))
mean_hoursepower_type.plot(kind='bar')
plt.title('Mean horsepower by car type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:** No
